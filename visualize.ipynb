```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy import stats

df = pd.read_csv('csv/50.csv', delimiter=',')
df.head()
min = df['GOP Majority'].min()
max = df['GOP Majority'].max()
```

```python
def normalize_min_max(min, max):
    if ((min > 0) or (max < 0)):
        return -1, -1
    if (abs(min) < max):
        return 0-max, max
    if (max < abs(min)):
        return min, 0-min
    return min, max

(n_min, n_max) = normalize_min_max(min, max)
```

```python
sns_plot = sns.relplot(x="Dem/Ind Prob", y="GOP Prob", palette="RdBu_r", hue="GOP Majority", hue_norm=(n_min, n_max), data=df)
plt.title('Age 50')
sns_plot.savefig('output.png')
```
